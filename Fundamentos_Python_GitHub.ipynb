## Operadores aritméticos

      Utilizados na execução de operações matemáticas.
        - '+': Soma.
        - '-': Subtração.
        - '*': Multiplicação.
        - '/': Divisão.
        - '//': Divisão inteira. A parte fracionada (ou resto) é descartado de forma automática.
        - '%': Módulo. Retorna o resto da divisão entre dois números.
        - '**': Potenciação. Retorna o valor de um número elevado à potência de outro.


---




```python
# Adição ou soma
n1 = 5.784
n2 = 7.17
soma = n1 + n2
print(f"A soma dos números é: {soma:.2f}.")
print (type(soma))
```

```python
# Subtração
a1 = 184.489
a2 = 89.741
sub = a1 - a2
print(f"O valor da subtração entre os números é: {sub:.2f}.")
print (type(sub))
```

```python
# Multiplicação
m1 = 4.65
m2 = 7

multi = m1 * 3
multi2 = m2 * m1

print ("O resultado da multiplicação da variável 'multi' é:", round (multi,2)) # round: Rretorna o valor fornecido em multi arredondado com a precisão de numa quantidade x de dígitos (no exemplo, 2 dígitos) após o ponto decimal.
print (f"A multiplicação dos números é: {multi2:.2f}.")
```

```python
# Divisão | Divisão inteira | Resto da divisão
d1 = 80
d2 = 24

div = d1 / d2
div_int = d1 // d2
div_resto = d1 % d2
print(f"Os números a serem divididos são: {d1} e {d2}.")
print("==========================================================================")
print (f"O resultado da DIVISÃO entre os números é: {div:.2f}.")
print("==========================================================================")
print (f"O resultado da DIVISÃO INTEIRA entre os números é: {div_int}")
print("==========================================================================")
print (f"O RESTO DA DIVISÃO entre os números é: {div_resto}")
```

## Operadores relacionais (comparação)



---

Foto 4.png

    Utilizados para exprimir uma relação entre valores, retornando um resultado booleano: TRUE ou FALSE.

    O uso desses operadores permite a criação de condições nos códigos.



```python
a = 8
b = 8
c = 5.14
d = -1

expressao1 = a == b # Igualdade
expressao2 = b != a # Diferença
expressao3 = c > d  # Maior que
expressao4 = d >= c # Maior ou igual a
expressao5 = d < a  # Menor que
expressao6 = c <= a # Menor ou igual a

print (f"Os valores a serem comparados, ou relacionados, são:\na = {a}\nb = {b}\nc = {c}\nd = {d}\n")
print(f"A primeira expressão diz que 'a' é igual a 'b' e vai retornar um valor '{expressao1}'.")
print("==========================================================================")
print(f"A segunda expressão diz que 'b' é diferente de 'a' e vai retornar um valor '{expressao2}'.")
print("==========================================================================")
print(f"A terceira expressão diz que 'c' é maior que 'd' e vai retornar um valor '{expressao3}'.")
print("==========================================================================")
print(f"A quarta expressão diz que 'd' é maior ou igual a 'c' e vai retornar um valor '{expressao4}'.")
print("==========================================================================")
print(f"A quinta expressão diz que 'd' é menor que 'a' e vai retornar um valor '{expressao5}'.")
print("==========================================================================")
print(f"A sexta expressão diz que 'c' é menor ou igual a 'a' e vai retornar um valor '{expressao6}'.")
print("==========================================================================")
```

## Operadores lógicos



---
    AND: Retorna TRUE se as duas CONDIÇÕES forem VERDADEIRAS
    OR: Retorna TRUE se UMA das CONDIÇÕES for VERDADEIRA.
    NOT: Inverte o resultado lógico.


### AND

```python
teste = 3 == 2 and 3 != 2
teste1 = 3 > 2 and 3 != 2

print(f"Na variável 'teste', somente uma das condições é verdadeiras, retornando um resultado '{teste}'.")
print(f"Na variável 'teste1', as duas condições são verdadeiras, retornando um resultado '{teste1}'.")
```

### OR

```python
teste2 = 3 >= 2 or 3 == 2 or 3 > 4

print(f"Na variável 'teste2', pelo menos uma das condições é verdadeiras, retornando um resultado '{teste2}'.")
```

### NOT

```python
teste3 = not(teste2)
print(f"Ao invertermos o resltado de 'teste2' através do uso do not, temos o seguinte retorno: '{teste3}'.")


teste4 = ((not(3 <= 2))) & (not(3 == 2)) & (3 < 4) # invertendo o resultado da primeira sentença e da segunda sentença
#               V                 V            V
print(f'((not(3 <= 2))) & (not(3 == 2)) & (3 < 4) é: {teste4}.')
teste5 = not((3 >= 2) & (3 != 2) & (3 < 4)) # invertendo o resultado da comparação de todas as sentenças
#                 V         V           V
print(f'not((3 >= 2) & (3 != 2) & (3 < 4)) é: {teste5}')
print(teste5)
```

## Operadores de identidade e de associação

---

    IS - Retorna TRUE se as variáveis comparadas FOREM O MESMO OBJETO (valor e tipo).
    IS NOT - Retorna TRUE se as variáveis comparadas NÃO FOREM O MESMO OBJETO (valor e tipo).

    IN - Retorna TRUE se um caracter ou uma sequencia estiver em uma str.
    NOT IN - Retorna verdadeiro se o caracter ou sequencia não estiver em uma str.

### IS

```python
h = 3
p = 3
teste_id = p == h
print(f'p == h é: {teste_id}')
teste_id1 = p is h
print(f'p is h é: {teste_id1}')
print("=================================================")
m = 3
n = 3.0
teste_id2 = m == n
print(f'm == n é: {teste_id2}')
teste_id3 = m is n
print(f'm is n é: {teste_id3}')
print(type(m))
print(type(n))
```

### IS NOT

```python
w = 3
z = 4
sem_op_id = w != z
com_op_id = w is not z
print(f"A variável 'w' é igual a {w}.\nA variável 'z' é igual a {z}.")
print("=================================================")
print(type(w))
print(type(z))
print("=================================================")
print(f"A primeira expressão não usa operador de identidade, e diz que 'w' É DIFERENTE de 'z', resultando em um {sem_op_id}.")
print(f"A segunda expressão usa operador de identidade, e diz que 'w' NÃO É 'z', resultando em um {sem_op_id}.")

```

## Estruturas condicionais

---

    IF - ELIF - ELSE
    É um bloco de código iniciado com uma expressão para avaliar se uma determinada condição é verdadeira ou falsa.
    O comando é executado de acordo com as condiçoes estabelecidas



---


### IF

    Executa um bloco de código se determinada condição for verdadeira.
    SINTAXE:
    if condição:
      Algo a ser executado se a condição for verdadeira

```python
idade = 20
if idade >= 18:
  print("Você é maior de idade.")
```

### ELSE

    Especifica um bloco de código a ser executado se a condição IF não for verdadeira.
    SINTAXE:
    if condição:
      Algo a ser executado se a condição for verdadeira
    else:
      Executa algo quando após a condição 'if' não for verdadeira


```python
idade = int(input("Qual sua idade? "))
if idade >= 18:
  print("Você é maior de idade.")
else:
  print("Você ainda não é maior de idade.")
```

### ELIF

    Abreviação de 'else if' e permite testar múltiplas condições em sequência.
    SINTAXE:
    if condição:
      Algo a ser executado se a condição for verdadeira.
    elif 2ª condição:
      Algo a ser executado se a 2ª condição for verdadeira.
    else:
      Algo a ser executado se nenhuma das condições anteriores for verdadeira.
    

```python
nota = float(input("Qual foi sua nota? "))

if nota >= 80:
  print("Parabéns. Você foi aprovado.")
elif nota >= 60 and nota < 80:
  print("Você está em recuperação. Estude um pouco mais.")
else:
  print("Que pena! Você foi reprovado.")
```

```python
# 1. Solicite ao usuário que digite um número.
# 2. Armazene o número digitado em uma variável chamada "numero".
# 3. Verifique se o número digitado é igual a zero.
  #  Se sim, exiba a mensagem "O número digitado é zero."
# 4. Caso contrário, verifique se o número é maior que zero.
  #  Se sim, exiba a mensagem "O número digitado é positivo."
# 5. Se as condições anteriores não forem atendidas, então o número é negativo.
  #  Nesse caso, exiba a mensagem "O número digitado é negativo."
# 6. Encerre o programa.

numero = float(input("Digite um número: "))

if numero == 0:
  print("o número digitado é zero.")
elif numero > 0:
  print("O número digitado é positivo.")
else:
  print("O número digitado é negativo.")
```

```python
# Objetivo: percorrer toda a lista de notas e para cada valor encontrado testar se o mesmo é ou não maior ou igual a 7
# Se verdadeiro imprima a mensagem aprovado e sua nota
# Se falso imprima reprovado e sua nota
notas = [5, 9, 6, 8, 7]
for i, v in enumerate(notas):
  if notas[i] >= 7: # se uma nota qualquer encontrada for maio ou igual a 7 faça algo
    print(f'aprovado com nota {notas[i]}.') # se verdadeiro imprima aprovado e sua nota
  else:
    print(f'reprovado com nota {notas[i]}.') # se não for verdadeiro imprima reprovado e sua nota
  print(notas[i])
```

```python
# Elabore uma função que receba em seis inputs diferentes as seguintes informações:
#   - Receita líquida em R$
#   - Custo em R$
#   - Estoques em R$
#   - Contas a receber em R$
#   - Fornecedores em R$
#   - O Número de dias do balanço (int)
# Após receber as informações acima a função deve calcular o valor e printá-lo dos seguintes KPIs:
#   - Prazo médio de estoques - PME
#   - Prazo médio de Recebimento - PMR
#   - Prazo médio de Pagamento PMP
#   - Ciclo operacional - CO
#   - Ciclo de caixa CC
# E por último:
# - Se PMP for maior que ciclo operacional, print a mensagem: A empresa paga em média tantos dias depois de receber
# - Se PMP for menor que ciclo operacional, print a mensagem: A empresa paga em média tantos dias antes de receber
# - Se PMP for igual ao ciclo operacional, print a mensagem: A empresa paga em média no mesmo dia em que recebe

def indicadores_financeiros():
  receita_liquida = float(input("Insira a receita líquida em reais: R$"))
  custo = float(input("Insira o custo em reais: R$"))
  estoque = float(input("Insira o estoque em reais: R$"))
  contas_a_receber = float(input("Insira as contas a receber em reais: R$"))
  fornecedores = float(input("Insira o valor de fornecedores em reais: R$"))
  dias = int(input("Insira a quantidade de dias (sendo um mês 30 dias): "))

  pme = (estoque / custo) * dias
  pmr = (contas_a_receber / receita_liquida) * dias
  pmp = (fornecedores / custo) * dias
  co = pme + pmr
  cc = co - pmp # cc = ciclo de caica

  if pmp > co:
    print(f"Essa empresa paga, em média, {cc:.1f} dias depois de receber.")
  elif pmp < co:
    print(f"Essa empresa paga, em média, {cc:.1f} dias antes de receber.")
  else:
    print(f"Essa empresa paga, em média, no mesmo dia em que recebe.")

  if cc < 0:
    print("==================================================================================")
    print(f"Caso ciclo de caixa negativo, significa que a empresa recebe os itens antes de pagar.")
indicadores_financeiros()
```

```python
# Faça uma função sem parâmetro que ao ser acionada solicite ao usuário uma sequência númerica
# A sequencia deve estar em uma única caixa de texto e separada por espaços
# A função tem que devolver um print informando o valor da média, variância e desvio padrão da sequência informada.

def medidas_estatisticas():
  seq_num = input("Digite uma sequencia de 05 números, separados por espaços: ")
  lista_num = []
  lista2_num = seq_num.split() # Transforma os dados digitaddos pelo usuário em uma lista
                               # Quando se usa o '.split()', os itens da lista ficam todos como tipo string.
  lista_num = []

  # Fazendo a tipagem dos itens.
  for i in lista2_num:    # "Para cada item da lista2_num, faça algo."
    if type(i) != float:  # Se o tipo do item for diferente de float
      x = float(i)        # Faça a tipagem do item para float.
      lista_num.append(x) # Usando o append, anexe cada 'x' na lista_num, já como float.

  # Média é a soma dos valores da sequencia, dividido pelo número de registros da sequencia.
  media = sum(lista_num) / len(lista_num)
  print("=======================================================")
  print(f'A média dos valores é: {media:.1f}')

  # Variância é que a cada item da sequencia, diminui-se da média, e eleva ao quadrado.
  # Dividindo tudo isso pelo valor do número de registros - 1.
  contador = []
  for i in lista_num:
    contador.append((i - media) ** 2)
  total = sum(contador)
  variancia = total / (len(lista_num) - 1)
  print("=======================================================")
  print(f"A variância amostral da sequencia é: {variancia:.1f}.")

  # Desvio padrão é o resultado da raiz quadrada da variância
  dp = variancia ** (1 / 2)

  print("=======================================================")
  print(f"O desvio-padrão desta amostra é de: {dp:.1f}.")

medidas_estatisticas()
```

## Loop While

      Utilizada quando se quer que determinado bloco de código seja executado ENQUANTO determinada condição for satisfeita.

      Sintaxe:
      while(condição)
        código


    No while, a condição pode ser reduzida à True ou False.

      - A verificação de valor de uma variável;
      - Determinada estrutura alcançar um tamanho;
      - O retorno de uma função se igualar a determinado valor;


```python
c = 0
n = int(input('Informe quantas vezes deseja ver a mensagem "Hello World": '))
while c < n: # Enquanto o contador for menor que o número digitado pelo usuário, faça algo.
  print('Hello World')
  c += 1 # contador = contador + 1

# O while é uma sentença que declara uma condição que, enquanto verdadeira
# executa repetidamente um bloco de códigos na linha por linha na ordem
# especificada de cada linha.
```

```python
# Faça um script que receba do usuário n números inteiros.
# Armazene os números recebidos em uma lista;
# Ao final exiba a soma desses valores.
# O usuário deverá informar apenas 1 numero inteiro por caixa de diálogo

contador = 0 # criei um contador para utilizar como parada do laço while
n = int(input('Informe a qtde de números inteiros que deseja na lista: '))
numeros = []
while contador < n:
  numeros.append(int(input(f'Informe o {contador+1}º número inteiro da lista: ')))
  contador += 1
print(f'A soma dos valores digitados é {sum(numeros)}')
```
